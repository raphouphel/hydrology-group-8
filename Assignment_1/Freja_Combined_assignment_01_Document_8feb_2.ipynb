## 2.) Transient water balance
The equation for transient water transportation is:
$$
dStor/dt=P*A-ET*A-Q-GWE
$$
Where P is precipitation, ET is evapotranspiration, A is area of catchment Q is catchment outflow and GWE is groundwater exchange.

In this assignmen we will not be working with GWE as it is assumed that all GWE will end up in the river near to cathments and that they are therefore already included in Q.

## 3) Long term average water balance
We assume that the catchments are steady-state and therefore the longterm change in water storage will equal 0. This means on a long term avererage the equation mentioned in previous task will look like this:

$$
P_{avg}*A-ET_{avg}*A-Q_{avg} = 0
$$

## 4) long term actual ET
We can isolate ET in from the long term storage equation:
$$
ET_{avg} = P_{avg}-Q_{avg}/A
$$

```python
# Loading data for catchment 1 to 4
# mean of precipitation
p1_avg = st1.precipitation.mean() #mm/d
p2_avg = st2.precipitation.mean() #mm/d
p3_avg = st3.precipitation.mean() #mm/d
p4_avg = st4.precipitation.mean() #mm/d
# mean of observed streamflow
Q1_avg = float(st1.Qobs.mean() * 1000 * (3600*24)) #mm*m2/d converted from m^3/s
Q2_avg = float(st2.Qobs.mean() * 1000 * (3600*24)) #mm*m2/d
Q3_avg = float(st3.Qobs.mean() * 1000 * (3600*24)) #mm*m2/d
Q4_avg = float(st4.Qobs.mean() * 1000 * (3600*24)) #mm*m2/d
print(Q1_avg,Q2_avg,Q3_avg,Q4_avg)
# area
A1 = 6463 * 10000 #m2
A2 = 62278 * 10000 #m2
A3 = 27048 * 10000 #m2
A4 = 11211 * 10000 #m2
# Actual evapotranspiration calculated from streamflow and area
ET1_avg = p1_avg - Q1_avg/A1
ET2_avg = p2_avg - Q2_avg/A2
ET3_avg = p3_avg - Q3_avg/A3
ET4_avg = p4_avg - Q4_avg/A4

print("p_avg for catchment 1 is " + str(round(p1_avg,2)))
print("p_avg for catchment 2 is " + str(round(p2_avg,2)))
print("p_avg for catchment 3 is " + str(round(p3_avg,2)))
print("p_avg for catchment 4 is " + str(round(p4_avg,2)))

print(ET1_avg,ET2_avg,ET3_avg,ET4_avg)
```

## 8) Reformulate the transient water balance question from question 2 for discrete time steps of 1 day duration.
$$Stor_{i+1} =Stor_{i} + (P_{i}*A_{i}-ET_{i}*A-Q_{i})* \Delta t$$

Here, $\Delta t$ indicates the length of time periods, e.g. daily or monthly for which instances of values with i is counted. This means it that if $\Delta t$ equals a day then $P_{i}$ is the precipitation for a whole day.

```python
#Areas
A1 = 6463 * 10000 #m2
A2 = 62278 * 10000 #m2
A3 = 27048 * 10000 #m2
A4 = 11211 * 10000 #m2
# FREJAS VERSION OF ESTIMATE FUNCTION
n_days=len(st1)

# Catchment 1
def qest_1(stor_0, alpha, beta):
    P = st1.precipitation.mul(10**(-3)) # mm/d to m/d
    ET_ref = st1.pet.mul(10**(-3)) # mm/d to m/d
    ET = ET_ref.mul(alpha) #m/d
    stor_ar = [0]*n_days

    discharge_estimate = [0]*n_days
    for i in range(n_days):
        if i == 0:
            stor_prev = stor_0*A1 # unit metres changes to m^3
        else:
            stor_prev = stor_ar[i-1]
        discharge_estimate[i] = beta*stor_prev
        Q_out = beta*stor_prev
        stor_ar[i] = stor_prev + (P[i]-ET[i])*A1 - Q_out
        if stor_ar[i] < 0:
            stor_ar[i] = 0
    discharge_estimate = pd.Series(discharge_estimate)
    avg = sum(discharge_estimate) / len(discharge_estimate)

    return avg

# Catchment 2
def qest_2(stor_0, alpha, beta):
    P = st2.precipitation.mul(10**(-3)) #m/d converted from mm/d
    ET_ref = st2.pet.mul(10**(-3)) #m/d converted from mm/d
    ET = ET_ref.mul(alpha) #m/d
    stor_ar = [0]*n_days

    discharge_estimate = [0]*n_days
    for i in range(n_days):
        if i == 0:
            stor_prev = stor_0*A2
        else:
            stor_prev = stor_ar[i-1]
        discharge_estimate[i] = beta*stor_prev
        Q_out = beta*stor_prev
        stor_ar[i] = stor_prev + (P[i]-ET[i])*A2 - Q_out
        if stor_ar[i] < 0:
            stor_ar[i] = 0
    discharge_estimate = pd.Series(discharge_estimate)
    avg = sum(discharge_estimate) / len(discharge_estimate)

    return avg

# Catchment 3
def qest_3(stor_0, alpha, beta):
    P = st3.precipitation.mul(10**(-3)) #m/d converted from mm/d
    ET_ref = st3.pet.mul(10**(-3)) #m/d converted from mm/d
    ET = ET_ref.mul(alpha) #m/d
    stor_ar = [0]*n_days

    discharge_estimate = [0]*n_days
    for i in range(n_days):
        if i == 0:
            stor_prev = stor_0*A3
        else:
            stor_prev = stor_ar[i-1]
        discharge_estimate[i] = beta*stor_prev
        Q_out = beta*stor_prev
        stor_ar[i] = stor_prev + (P[i]-ET[i])*A3 - Q_out
        if stor_ar[i] < 0:
            stor_ar[i] = 0
    discharge_estimate = pd.Series(discharge_estimate)
    avg = sum(discharge_estimate) / len(discharge_estimate)

    return avg

# Catchment 4
def qest_4(stor_0, alpha, beta):
    P = st4.precipitation.mul(10**(-3)) #m/d converted from mm/d
    ET_ref = st4.pet.mul(10**(-3)) #m/d converted from mm/d
    ET = ET_ref.mul(alpha) #m/d
    stor_ar = [0]*n_days

    discharge_estimate = [0]*n_days
    for i in range(n_days):
        if i == 0:
            stor_prev = stor_0*A4
        else:
            stor_prev = stor_ar[i-1]
        discharge_estimate[i] = beta*stor_prev
        Q_out = beta*stor_prev
        stor_ar[i] = stor_prev + (P[i]-ET[i])*A4 - Q_out
        if stor_ar[i] < 0:
            stor_ar[i] = 0
    discharge_estimate = pd.Series(discharge_estimate)
    avg = sum(discharge_estimate) / len(discharge_estimate)

    return avg

# Catchment
print(qest_1(10,0.3,0.004))

# Initial parameters
stor_ini = 10 #mm
alpha_ini = 0.3
beta_ini = 0.004 #1/d

```
